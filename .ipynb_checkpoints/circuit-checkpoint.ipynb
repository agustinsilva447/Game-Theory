```python
import qiskit
import sympy as sp
import numpy as np
from qiskit import QuantumCircuit, QuantumRegister
from qiskit.quantum_info import Operator
```

```python
theta1 = qiskit.circuit.Parameter('theta1')
phi1 = qiskit.circuit.Parameter('phi1')
lam1 = qiskit.circuit.Parameter('lam1')

theta2 = qiskit.circuit.Parameter('theta2')
phi2 = qiskit.circuit.Parameter('phi2')
lam2 = qiskit.circuit.Parameter('lam2')

circ = QuantumCircuit(1)
circ.u(theta1,phi1,lam1,0)
circ.id(0)
circ.u(theta2,phi2,lam2,0)
circ.draw('mpl', scale = 2)
```
